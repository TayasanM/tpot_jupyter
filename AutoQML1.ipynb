```python
import pandas as pd
from datetime import timedelta
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

from autoqml import TabularClassification, AutoQMLFitCommand

# data
X, y = load_iris(return_X_y=True, as_frame=True)  # pandas DataFrame + Series
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=777, stratify=y
)

# optional: keep search space small for a quick test
custom_config = {
    # Example: force a classical algorithm first (fast sanity check)
    "autoqml.search_space.classification.ClassificationChoice__choice": "svc",
    "autoqml.search_space.classification.ClassificationChoice__choice": "qsvc",
    "autoqml.search_space.preprocessing.rescaling.RescalingChoice__choice": "min_max_scaling",
    "autoqml.search_space.preprocessing.dim_reduction.DimReductionChoice__choice": "no-op",
    "autoqml.search_space.classification.quantum.qsvc.QSVC__num_repetitions": 2,
    "autoqml.search_space.classification.quantum.qsvc.QSVC__num_qubits": 2,
    "autoqml.search_space.classification.quantum.qsvc.QSVC__C": 10,
}

autoqml_pipeline = TabularClassification()

cmd = AutoQMLFitCommand(
    X_train,
    y_train,
    timedelta(seconds=120),     # small budget for quick run
    configuration=custom_config # can be {} for full search
)

autoqml_pipeline.fit(cmd)

y_pred = autoqml_pipeline.predict(X_test)

from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(y_test, y_pred))
```
